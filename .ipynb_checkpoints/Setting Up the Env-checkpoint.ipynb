## Setting the API Key as an Environment Variable

```python
import os
```

```python
os.environ.items()
```

```python
for key, value in os.environ.items():
    print(f"{key}:{value}")
```

```python
os.environ['OPENAI_API_KEY'] = ''
```

```python
%load_ext dotenv
%dotenv
```

```python
for key, value in os.environ.items():
    if key == 'OPENAI_API_KEY':
        print(f"{key}:{value}")
```
