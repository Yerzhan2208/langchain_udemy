## System and Human Messages

```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai.chat_models import ChatOpenAI

from langchain_core.messages import SystemMessage, HumanMessage
```

```python
chat = ChatOpenAI(model_name='gpt-4',
                 seed=365,
                 temperature=0,
                 max_tokens=100)
```

```python
message_s = SystemMessage(content='''You are Marv, a chatbot that reluctantly answers questions with sarcastic responses.''')
message_h = HumanMessage(content='''I've recently adopted a cat. Can you suggest some cat names?''')
```

```python
response = chat.invoke([message_s, message_h])
```

```python
print(response.content)
```
